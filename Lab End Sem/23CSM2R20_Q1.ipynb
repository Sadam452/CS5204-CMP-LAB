```python
y = data['MedHouseVal'], 
X = data.drop('MedHouseVal', axis=1)
```

```python
columns = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup',
       'Latitude', 'Longitude', 'MedHouseVal']
```

```python
#center the data by subtracting the mean
#1. Center the data by subtracting the mean from each feature column.
for i in range(0, len(columns)):
    data[columns[i]] -= np.mean(data[columns[i]])
```

```python
data.describe()
```

```python
#Calculate the covariance matrix of the centered data.
data_t = np.transpose(data)
cov = np.cov(data_t)
```

```python
print(cov.shape)
```

```python
cov
```

```python
#. Find the eigenvalues and eigenvectors of the covariance matrix.
eigenvalues, eigenvectors = np.linalg.eig(cov)
```

```python
print("Eigenvalues:",eigenvalues)
print("Eigenvectors:",eigenvectors)
```

```python
#4. Select the top k eigenvectors corresponding to the largest eigenvalues (where
#kÂ¡ d is the desired reduced dimension).
k=6
vec = np.transpose(eigenvectors[:k])
data2 = np.dot(data, vec)
```

```python
data2 = pd.DataFrame(data2)
data2
```

```python
for i in range(0,5):
    plt.scatter(data2[i], y)
    plt.title(i)
    plt.show()
```
